```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

low_contrast = mpimg.imread('../imagens/Einstein_low_contrast.png')
med_contrast = mpimg.imread('../imagens/Einstein_med_contrast.png')
high_contrast = mpimg.imread('../imagens/Einstein_high_contrast.png')
```

```python
fig = plt.figure(figsize=(15,12))
a = fig.add_subplot(1, 3, 1)
plt.imshow(low_contrast)
a.set_title('Low Contrast')
a = fig.add_subplot(1, 3, 2)
plt.imshow(med_contrast)
a.set_title('Medium Contrast')
a = fig.add_subplot(1, 3, 3)
plt.imshow(high_contrast)
a.set_title('High Contrast')
```

```python
fig = plt.figure(figsize=(15,3))

a = fig.add_subplot(1, 1, 1)
a.hist(low_contrast.ravel(), bins=256, range=(0.0, 1.0))
a.set_title('Low Contrast')

a = fig.add_subplot(1, 3, 2)
a.hist(med_contrast.ravel(), bins=256, range=(0.0, 1.0))
a.set_title('Medium Contrast')

a = fig.add_subplot(1, 3, 3)
a.hist(high_contrast.ravel(), bins=256, range=(0.0, 1.0))
a.set_title('High Contrast')

plt.show()
```
